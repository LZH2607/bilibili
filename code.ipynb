```python
import json
import pandas as pd
from utils import *
```

```python
raw_data_list = []
```

```python
for i in range(1, 43):
    directory = "json"
    file_name = str(i)
    file_path = directory + "/" + file_name
    json_string = load_text_to_string(file_path)
    jzon = json.loads(json_string)
    num = len(jzon['data']['list'])
    for j in range(0, num):
        raw_data = jzon['data']['list'][j]
        raw_data_list.append(raw_data)
```

```python
len(raw_data_list)
```

```python
data_list = []
```

```python
for raw_data in raw_data_list:
    data = []
    data.append(raw_data['mid'])
    data.append(raw_data['uname'])
    data.append(raw_data['face'].replace(" ", ""))
    data.append(raw_data['sign'])
    data.append(raw_data['official_verify']['desc'])
    data.append(str(raw_data))
    data.append
    # print(data)
    data_list.append(data)
```

```python
for data in data_list:
    print(data)
```
